## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
# Your code here
def count_lines (text):
    return len(text)

df['numberlines'] = df.linename.map(count_lines)

#df['numberlines'] = df.linename.map (lambda number : len (number))

#df.numberlines.sort_values(ascending = False)
df.sort_values('numberlines', ascending = False)
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
sumdays = sumdays.reset_index()
sumdays.head()
sumdays['isweekend'] = sumdays.dayoftheweek.map ({0:False, 1:False, 2:False, 3:False, 4:False, 5:True, 6:True})
weekend = sumdays.groupby('isweekend').mean()
weekend [['entries','exits']].plot(kind = 'barh')
```
